```python
import pandas as pd
import numpy as np
import matplotlib as matplot
import matplotlib.pyplot as plt
from os import listdir
from os.path import isfile, join
from IPython.display import display, HTML
import seaborn as sns
```

```python
pd.set_option('display.max_rows', 200)
```

Questa pagina ha usato come base di dati i dati disponibilizzati sul [questa](https://www.reggionline.com/elezioni-amministrative-reggio-emilia-sara-sindaco-reggio-risultati-diretta/) notizia del giornale Reggio Online.
Ogni riga della tabella sulla pagina apre una nuova pagina con i voti del candidato in ogniuna delle 160 sezioni.




```python
onlyfiles = [f for f in listdir() if isfile(join("./", f)) and f.find(".csv") != -1 and f!= "cap.csv"]
onlyfiles.sort()
```

Sotto la lista dei candidati:

```python
candidates = [ f[:-4].capitalize() for f in onlyfiles]

print(candidates)
```

Qui si inizia la gestione dei dati.

```python
merged = pd.DataFrame()
```

Qui vengono stampati i voti di tutti i candidati, come presentati nel sito di Reggio Online.
In questo pezzo di codice Ã¨ fatto anche il merge delle tabelle, in modo che ci siano i voti di tutti i candidati divisi per sezione.

```python
display(HTML('<h2>- Risultati per sezione -</h2>'))
for f in onlyfiles:
    csv = pd.read_csv(f, delimiter=";",names=["Sezione", "Indirizzo", f[:-4].capitalize(), "%"], dtype={'Sezione': 'string', f[:-4].capitalize(): 'int64'})
    csv = csv.drop(["%"], axis=1)
    display(HTML(f"<h3> Candidato: {f[:-4].capitalize()}  </h3>"))
    display(csv)
    if merged.empty:
        merged = csv.copy()
    else:
        merged = merged.merge(csv, on=["Sezione", "Indirizzo"], how="inner").copy()
```

Il tabellone finale

```python
display(merged)
```

La lista con gli indirizzi di tutte le sezioni.

```python
merged["Indirizzo"].unique()
```

Qui Ã¨ stata fatta una raccolta dei CAP di tutti gli indirizzi sopra. I dati sono stati presi sull'internet, quindi Ã¨ possibile che ci siano degli errori.

```python
csv = pd.read_csv("cap.csv", delimiter=";",names=["Indirizzo", "CAP"], dtype={'CAP': 'string'})
```

I CAP sono inseriti nel tabellone principale.

```python
merged = merged.merge(csv, on="Indirizzo", how="left").copy()
```

Qui abbiamo il totale di voti dei candidati.

```python
merged.loc["Total"] = merged.sum(axis=0, numeric_only=True)
```

```python
display(merged)
```

In questo punto Ã¨ fatto un cambio della posizione della collona CAP.

```python
columns = merged.columns.values
first = np.append(columns[0:2], columns[-1])
newColumns = np.append(first, columns[2:len(columns)-1])
merged = merged.reindex(columns= newColumns)
```

Ora si fa una somma dei voti per CAP.

```python
lastCandidate = newColumns[-1]
groupedByCAP = merged.iloc[:, 2:].sort_values(by=["CAP"], ascending=True, ignore_index=True)
groupedByCAP = groupedByCAP.groupby(by="CAP").sum()
```

```python
print("Totale di voti per CAP")
display(groupedByCAP)
```

E anche il totale di elettori in ogni CAP.

```python
row_sums = groupedByCAP.sum(axis=1)
elettoriPerCAP = pd.DataFrame(data={'CAP': row_sums.index, 'Elettori': row_sums.values})
display(elettoriPerCAP)
row_sums.plot(kind='bar')
```

Il CAP 42123 (Reggio Emilia Est) Ã¨ quello con piÃ¹ elettori. Il CAP 42121 Ã¨ quello con meno elettori.

```python
percentage_df = groupedByCAP.div(row_sums, axis=0) * 100
percentage_df = percentage_df.round(2)
print("Percentuale per CAP")
display(percentage_df)
```

Non sono sicuro di quanto questo mappa sia esatto, ma sembra maggiormente corretto.

caps.jpg

Un semplice paragoni tra i due candidati piÃ¹ votati, per CAP. Qui vediamo che la differenza tra i due candidati Ã¨ stata sempre superiore a 20 punti percentuali, eccetto su centro.

```python
capList = percentage_df.index
diffMassariTarchini = percentage_df['Massari'] - percentage_df['Tarchini']
diffByCAP = pd.DataFrame({'Differenza': diffMassariTarchini})
display(diffByCAP)
```

Grafici di votazione assoluta per CAP.

```python
# Plotting
for index, row in groupedByCAP.iterrows():
    plt.figure()
    row = row.sort_values(ascending=False)
    row.plot(kind='bar')
    plt.title(f'CAP {index}')
    plt.ylabel('Values')
    plt.xlabel('Categories')
    plt.show()
```

Votazione di ogni candidato per CAP.

```python
ax = percentage_df.transpose().plot.barh()
plt.show()
```

Sotto abbiamo un zoom sui candidati meno votati.

```python
ax = percentage_df.filter(items=["Tasselli", "Soragni", "Sabillon", "Reggio", "Aguzzoli"]).transpose().plot.barh()
plt.show()
```

Percentuale di voti per CAP. Tarchini Ã¨ stato piÃ¹ forte sul CAP con meno elettori. Tra i candidati minori non ci sono tanti cambiamenti.

```python
for index, row in groupedByCAP.iterrows():
    plt.figure()
    explode = [0.03 for i in range(0, len(candidates))]
    row.plot(kind='pie', explode=explode, radius=1.3, autopct='%1.1f%%', startangle=160, legend=False, labeldistance=1.25, pctdistance=1.1, textprops={'fontsize': 8})
    plt.title(f'CAP {index}', loc='left')
    plt.ylabel('')  # Hide the y-label for better appearance
    plt.show()
```

Le percentuali dei voti di ogni candidato diviso per CAP. Il CAP 42123 rappresenta quasi sempre la percentuali piÃ¹ grossa.
Un distacco per Sabillon, l'unico candidato di origine straniera, che ha preso 50% dei suoi voti alla zona della via Turri e quartiere Santa Croce.

```python
for index, row in groupedByCAP.transpose().iterrows():
    plt.figure()
    explode = [0.03 for i in range(0, len(capList))]
    row.plot(kind='pie', explode=explode, radius=1, autopct='%1.1f%%', startangle=90, legend=True, labeldistance=None, pctdistance=0.8, textprops={'fontsize': 8})
    plt.title(f'{index}', loc='left')
    plt.legend(labels=capList, bbox_to_anchor=(0.85,1.025), loc="upper left")
    plt.ylabel('')  # Hide the y-label for better appearance
    plt.show()
```
